## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['subtotal'] = df['unit_price'] * df['qty']
df
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

# Calculate the shipping price based on the weight of the items
df['shipping_price'] = df['unit_weight'].apply(lambda x: 7 if x > 50 else 10) * df['unit_weight']

# Display the updated dataframe
df
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

# Calculate the total price including sales tax
df['total_price'] = df['subtotal'] + df['shipping_price'] + (df['subtotal'] + df['shipping_price']) * 0.0925

# Display the updated dataframe
df


```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']

# Display the updated dataframe
df
```

```python
# Create a column for the profit of
# each line using line cost and line price
#df['line_profit'] = df['line_price'] - df['line_cost']

# Create a column for the profit of each line
df['line_profit'] = df['total_price'] - df['line_cost']


# Display the updated dataframe
df
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.

top_clients = df['client_id'].value_counts().head(5)
client_ids = top_clients.index.tolist()

client_ids

client_spending = df[df['client_id'].isin(client_ids)].groupby('client_id')['total_price'].sum()

# Sort the values in descending order

client_spending = client_spending.sort_values(ascending=False)
client_spending


# Format using Dollars

client_spending = client_spending.map('${:,.2f}'.format)
client_spending

```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

# Get the top 5 clients by quantity
top_clients = df['client_id'].value_counts().head(5)
client_ids = top_clients.index.tolist()

# Filter the dataframe for the top 5 clients
top_clients_df = df[df['client_id'].isin(client_ids)]

# Group the data by client_id and calculate the totals
summary_df = top_clients_df.groupby('client_id').agg({'qty': 'sum', 'shipping_price': 'sum', 'total_price': 'sum', 'line_profit': 'sum'})

# Sort the summary dataframe by total profit
summary_df = summary_df.sort_values('line_profit', ascending=False)

# Format the summary dataframe using Dollars

summary_df_formatted = summary_df.applymap('${:,.2f}'.format)
summary_df_formatted




```

```python
# Format the data and rename the columns
# to names suitable for presentation.

df_formatted = df.rename(columns={'first':'First', 'last':'Last', 'job':'Job', 'phone':"Phone", 'email':'Email', 'client_id': 'Client ID', 
                                  'order_id': 'Order ID', 'order_date':'Order Date', 'order_week':'Order Week', 'order_year':'Order Year', 
                                  'item_id':'Item ID', 'category':'Category', 'subcategory':'Subcategory', 'item_name':'Item Name', 'qty':'Quantity',
                                  'line_number':'Line Number','unit_price':'Unit Price', 'unit_cost':'Unit Cost', 'unit_weight':'Unit Weight', 'subtotal':'Subtotal',
                                  'shipping_price':'Shipping Price', 'total_price':'Total Price', 'line_cost':'Cost', 'line_profit':'Profit'})

# Currency should be in millions of dollars.
# Format the data to reflect that.

df_formatted['Unit Price'] = df_formatted['Unit Price'].map('${:,.2f}'.format)
df_formatted['Unit Cost'] = df_formatted['Unit Cost'].map('${:,.2f}'.format)
df_formatted['Subtotal'] = df_formatted['Subtotal'].map('${:,.2f}'.format)
df_formatted['Shipping Price'] = df_formatted['Shipping Price'].map('${:,.2f}'.format)
df_formatted['Total Price'] = df_formatted['Total Price'].map('${:,.2f}'.format)
df_formatted['Cost'] = df_formatted['Cost'].map('${:,.2f}'.format)
df_formatted['Profit'] = df_formatted['Profit'].map('${:,.2f}'.format)

# Display the formatted dataframe
df_formatted






```

```python
# Sort the updated data by "Total Profit" from highest to lowest

# Remove dollar signs and commas, then convert to float
df_formatted['Profit'] = df_formatted['Profit'].replace('[\$,]', '', regex=True).astype(float)

# Sort by 'Profit' in descending order
df_sorted = df_formatted.sort_values('Profit', ascending=False)
df_sorted['Profit'] = df_sorted['Profit'].map('${:,.2f}'.format)
df_sorted
```
